### Explanatory Data Analysis(EDA)

**Task 5: Data Analysis**

#### **Univariate Analysis**

```python
# Analysis of the price distribution
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='price',bins=100)
plt.title('Price Distriution')
plt.ylabel('Frequency')
plt.show()
```

The Histogram shows how rental prices are spread out across the Airbnb listings. The analysis indicated that approximately 70% of the listings were priced between $50 and $300, highlighting a significant segment of the rental market within this range. This insight suggests a competitive pricing strategy among hosts in this pricing bracket. Understanding these trends can provide valuable guidance for prospective hosts aiming to optimize their pricing strategies.

```python
# Analysing the availability_365 column distribution

plt.figure(figsize=(7,5))
sns.histplot(data=df, x='availability_365')
plt.title('Availability_365 Distriution')
plt.ylabel('Frequency')
plt.show()
```

The "Availability_365" histogram reveals a positively skewed distribution, with many listings having limited annual availability. This suggests that a significant portion of properties are only seasonally or part-time available, reflecting host preferences and seasonal demand patterns. These trends highlight diverse approaches to rental availability across New York’s Airbnb market.

#### Feature Engineering

```python
df.dtypes
```

```python
# Average price per neighbourhood_group

df.groupby('neighbourhood_group')['price'].mean()
```

The average price per neighborhood group reveals distinct pricing patterns across New York City:
- Manhattan: Highest average price at $204.15, reflecting its premium location and demand.<br>
- Brooklyn: Moderate pricing at $155.14, likely driven by its popularity and proximity to Manhattan.<br>
- Queens: Lower average price at $121.68, offering more affordable options.<br>
- Staten Island: Averages $118.78, reflecting its more residential and suburban appeal.<br>
- Bronx: Lowest average at $107.99, potentially due to lower tourist demand compared to other boroughs.<br>
These price variations highlight how location influences Airbnb pricing across New York City.<br>

```python
# Average price per bed by neighbourhood_group

df['price per bed'] = df['price']/ df['beds']

df.head()

```

```python
df.groupby('neighbourhood_group')['price per bed'].mean()
```

The average price per bed varies significantly across New York's neighborhood groups, indicating differing rental market dynamics. Manhattan has the highest average price per bed at approximately $138.71, suggesting a premium market often associated with high demand and limited supply. Brooklyn follows with an average of $99.79, reflecting its growing popularity among residents and tourists alike. Queens and the Bronx show moderate prices at around $76.34 and $74.71, respectively, which may appeal to budget-conscious renters. Staten Island has the lowest average price per bed at $67.73, potentially indicating less demand or different market characteristics. This data highlights the importance of location in rental pricing and can guide potential investors or renters in their decision-making processes.

#### Bivariate Analysis

```python
df.columns
```

```python
# Price dependency on neighbourhood_group

plt.figure(figsize=(8,6))
sns.barplot(data= df, x='neighbourhood_group', y= 'price', hue='room_type')
plt.title('Price Dependency on Neighbourhood_group')
plt.show()
```

The bar plot shows that Manhattan has the highest average Airbnb prices due to premium demand, followed by Brooklyn and Queens. The Bronx and Staten Island feature lower prices, reflecting budget-friendly options. Room type significantly affects pricing, with entire homes/apartments costing more than private and shared rooms. Hosts can optimize pricing strategies based on neighborhood and room type, while guests can choose accommodations based on budget and amenities. Understanding these trends aids decision-making for both hosts and guests in the Airbnb market.

```python
# number of reviews and price relationship
plt.figure(figsize=(8,6))
plt.title('Price and Review Dependency ')
sns.scatterplot(data= df, x= 'number_of_reviews', y= 'price', hue='neighbourhood_group' )
plt.show()
```

The scatter plot reveals a negative correlation between the number of reviews and Airbnb prices in New York City, indicating that listings with more reviews tend to have lower prices. Manhattan shows a wider price range for high-review listings, reflecting significant influences from location and amenities, while Brooklyn exhibits a similar but lower trend. Queens, Bronx, and Staten Island feature generally lower prices, likely due to reduced demand for budget-friendly options. Hosts can optimize pricing strategies based on review counts, while guests may use these metrics to assess value. Understanding these dynamics enhances decision-making in the Airbnb market.

```python
sns.pairplot(data=df , vars=['price','minimum_nights', 'number_of_reviews', 'availability_365'], hue= 'room_type')
```

The pairplot reveals key relationships among price, minimum nights, number of reviews, and availability in the Airbnb dataset. Price distribution is right-skewed, indicating most listings are affordable, while minimum nights and reviews also show similar skewness. Weak correlations suggest that higher prices may relate to longer minimum stays and fewer reviews. Room types indicate that entire homes are pricier, while shared rooms are the most budget-friendly. Overall, these insights highlight interconnected factors that can inform pricing strategies and enhance guest experiences.

```python
# Geographical Distribution Airbnb Listing
plt.figure(figsize=(10,7))
sns.scatterplot(data= df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical Distribution')
plt.show()
```

The scatter plot shows the geographical distribution of Airbnb listings in New York City, categorized by room type. Most listings are clustered in Manhattan, indicating high demand in this area. Entire homes/apartments dominate in Manhattan, while private rooms are prevalent in both Manhattan and Brooklyn. Hotel rooms are fewer and located near tourist hotspots, whereas shared rooms are scattered throughout the city. This distribution highlights the impact of location on guest preferences and can inform hosts about high-demand areas for strategic pricing and marketing.

```python
# Heat map - correlation of one variable with others for numerical columns

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8,6))
sns.heatmap(data= corr, annot= True)
```


The correlation heatmap highlights key relationships among numerical variables in the Airbnb dataset:

- **Price and Beds:** There is a positive correlation, suggesting that listings with more beds tend to have higher prices.
- **Price and Minimum Nights/Availability:** A negative correlation indicates that longer minimum stays and higher availability are associated with lower prices.
- **Reviews and Reviews per Month:** A strong positive correlation shows that listings with more reviews generally receive consistent monthly feedback.
- **Minimum Nights and Availability:** A weak positive correlation suggests that listings with longer minimum stays may have reduced overall availability.
- **Actionable Insights:** Hosts can utilize these correlations to refine pricing and marketing strategies, enhancing guest satisfaction and operational efficiency.
